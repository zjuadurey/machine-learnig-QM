```python
import json
import numpy as np
import math
import json
import matplotlib.pyplot as plt
```

```python
#StructureFactorX.real
#energy

stuXr = []
energy = []

for g in np.arange(0, 1.01, 0.05):
    data = json.load(open("log_data_6*6/<text%0.2f_iter_1000_6*6>.log"%(g)))
    print("g=%0.2f is loadded"%(g))
    enegry_each_g = []
    for iteration in data["StructureFactorX"]["Mean"]["real"]:
        enegry_each_g.append(iteration)
    enegry_each_g_mean = abs(np.mean(enegry_each_g[-20:]))
    energy.append(enegry_each_g_mean)
```

```python
energy
```

```python
x=np.arange(0, 1.01, 0.05)
plt.style.use('seaborn-darkgrid') # 设置画图的风格 
fig, ax = plt.subplots()
ax.plot(x, energy)
ax.set_xlabel('g')
ax.set_ylabel("StructureFactorX.real")
plt.title("6 x 6 ")
plt.show()
```
